```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("imagens/px-people.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
```

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_gray, cmap='gray')
```

```python
classifcador = cv2.CascadeClassifier(r"classificadores\haarcascade_frontalface_default.xml")
```

```python
faces = classifcador.detectMultiScale(imagem_gray, 1.3, 5)
```

```python
print(len(faces))
```

```python
faces[0]
```

```python
imagem_anotada = imagem.copy()
for (x, y, l, a) in faces:
    cv2.rectangle(imagem_anotada, (x, y), (x + l, y + a), (255, 0, 0), 2)

plt.imshow(imagem_anotada)
```

```python
face_imagem = 0

for (x, y, w, h) in faces:
    face_imagem += 1
    imagem_roi = imagem[y:y+h, x:x+w]
    imagem_roi = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
    cv2.imwrite("face"+ str(face_imagem)+ ".png", imagem_roi)
```
